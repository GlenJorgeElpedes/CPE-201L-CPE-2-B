```python
def display_array(arr):
    """Display all elements with their indices"""
    print("\nArray contents:")
    for idx in range(len(arr)):
        print(f"[{idx}] -> {arr[idx]}")

def show_reversed(arr):
    """Display the array in reverse order without modifying the original"""
    print("\nReversed array:", arr[::-1])

def update_value(arr, idx, new_val):
    """Update the value at a given index"""
    if 0 <= idx < len(arr):
        arr[idx] = new_val
        print(f"Updated index {idx} to {new_val}")
    else:
        print("⚠ Index out of range!")
    return arr

def arrange_array(arr):
    """Sort the array in ascending order"""
    arr.sort()
    print("\nArray sorted successfully.")
    return arr

def add_element(arr, idx, val):
    """Insert a value at a specific index"""
    arr.insert(idx, val)
    print(f"Inserted {val} at index {idx}")
    return arr

def remove_element(arr, idx):
    """Remove the element at the given index"""
    if 0 <= idx < len(arr):
        removed = arr.pop(idx)
        print(f"Removed {removed} from index {idx}")
    else:
        print("⚠ Index out of range!")
    return arr

def main():
    data = [10, 20, 30, 40, 50]
    print("Starting array:", data)

    while True:
        print("\n--- MENU ---")
        print("1. Traverse array")
        print("2. Reverse array")
        print("3. Modify array")
        print("4. Sort array")
        print("5. Insert array")
        print("6. Delete array")
        print("7. Exit")

        choice = input("Select an option (1-7): ")

        if choice == "1":
            display_array(data)
        elif choice == "2":
            show_reversed(data)
        elif choice == "3":
            try:
                idx = int(input("Enter index to update: "))
                new_val = int(input("Enter new value: "))
                data = update_value(data, idx, new_val)
                print("Updated array:", data)
            except ValueError:
                print("⚠ Please enter valid integers.")
        elif choice == "4":
            data = arrange_array(data)
            print("Sorted array:", data)
        elif choice == "5":
            try:
                idx = int(input("Enter index to insert at: "))
                val = int(input("Enter value to insert: "))
                data = add_element(data, idx, val)
                print("Array after insertion:", data)
            except ValueError:
                print("⚠ Please enter valid integers.")
        elif choice == "6":
            try:
                idx = int(input("Enter index to remove: "))
                data = remove_element(data, idx)
                print("Array after removal:", data)
            except ValueError:
                print("⚠ Please enter a valid integer.")
        elif choice == "7":
            print("Goodbye!")
            break
        else:
            print("⚠ Invalid choice, pick between 1 and 7.")

if __name__ == "__main__":
    main()
```
