```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.children = []

    def add_child(self, child_node):
        self.children.append(child_node)

    def remove_child(self, child_node):
        self.children = [child for child in self.children if child != child_node]

    def traverse(self):
        nodes = [self]
        while nodes:
            current_node = nodes.pop()
            print(current_node.value)
            nodes.extend(current_node.children)

    def delete_tree(self):
        """Deletes the tree using a post-order traversal."""
        # Recursively delete children first (post-order)
        for child in list(self.children): # Iterate over a copy to allow modification
            child.delete_tree()
        self.children.clear()  # Remove references to children from the current node
        print(f"Deleted node: {self.value}") # Optional: confirm deletion

    def __str__(self, level=0):
        ret = "  " * level + str(self.value) + "\n"
        for child in self.children:
            ret += child.__str__(level + 1)
        return ret

# Create a tree
root = TreeNode("Root")
child1 = TreeNode("Child 1")
child2 = TreeNode("Child 2")
grandchild1 = TreeNode("Grandchild 1")

grandchild2 = TreeNode("Grandchild 2")

root.add_child(child1)
root.add_child(child2)
child1.add_child(grandchild1)
child2.add_child(grandchild2)

print("Tree structure before deletion:")
print(root)

print("\nTraversal before deletion:")
root.traverse()

print("\nDeleting the tree:")
root.delete_tree()

print("\nTree structure after deletion (root should have no children and children should be reported as deleted):")
print(root)
print("Root children after deletion:", root.children)
```
